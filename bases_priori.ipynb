```python
dicionario = {len(abr):len(cabl), len(afi):len(cafi), len(bar):len(cbar), len(bja):len(cbja), len(cab):len(ccsa) , len(car):len(ccar), len(gar):len(cgar), len(iga):len(cigr),
              len(ipj):len(cipj), len(jab):len(cjbg), len(oli):len(coli), len(pes):len(cpes), len(plt):len(cplt), len(prm):len(cprm), len(rec):len(crec), len(rei):len(reipe),
              len(vsa):len(cvsa), len(vir):len(virtual)}
campus = ['CABL', 'CAFI', 'CBAR', 'CBJA', 'CCSA', 'CCAR', 'CGAR', 'CIGR', 'CIPJ', 'CJBG', 'COLI', 'CPES', 'CPLT', 'CPRM', 'CREC', 'REIPE', 'CVSA', 'VIRTUAL']
```

```python
print(dicionario)
```

```python
salas = pd.read_excel('priori/Anexo III rev2.xlsx', sheet_name=19)
```

```python
quant = pd.DataFrame(list(dicionario.items()), columns=['priori-06-out-2021', 'priori-29-nov-2021'])
quant['campi'] = campus
quant = quant[['campi',  'priori-06-out-2021', 'priori-29-nov-2021']]
quant['diferenca'] = quant['priori-06-out-2021'] - quant['priori-29-nov-2021']
total1 = quant['priori-06-out-2021'].sum()
total2 = quant['priori-29-nov-2021'].sum()
print(total1)
print(total2)

total = {'campi':'TOTAL', 'priori-06-out-2021':total1, 'priori-29-nov-2021':total2, 'diferenca':total1-total2}
quant = quant.append(total, ignore_index=True)

display(quant)
quant.to_excel('quantificacoes/quantificacao_priori.xlsx', index=False)
```
